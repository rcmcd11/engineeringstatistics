```python
import numpy as np #libraries
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20) #evenly spaced array between -10 and 10, with 20 total values
```

```python
x
```

```python
m = 50 #slope
b = 17 #intercept
```

```python
y = m * x + b #we can perform this function "y" on our array x and then plot it
```

```python
plt.plot(x,y,'.') #here is our x and y values plotted against each other (the function is linear)
```

```python
r = 100*np.random.randn(20,) #now we create a 20 random number list normally distributed around 1 and multiply by 100
```

```python
r
```

```python
plt.plot(r,'.') #plot our random numbers
```

```python
y2 = m * x + b + r #if we add our y function to these random numbers we can get a distribution to graph
```

```python
plt.plot(x,y2,'.')
```

```python
m_guess = 1 #we want a line which best approximates our data
b_guess = 1 #we can guess and see which line does

y_guess = m_guess * x + b_guess #we will guess m and b values and see which ones have the lowest "loss" of data

plt.plot(x,y_guess,'r'); #plot our guess line
plt.plot(x,y2,'g.'); #plot our data
```

```python
for m_guess in np.linspace(-100,100,100): #now we guess every slope between -100 and 100 (100 different slope values and see which fits)

        y_guess = m_guess * x #multiply our slope guess by the data to find it

        plt.plot(x,y_guess,'r'); #plot our guess line
        plt.plot(x,y2,'g.'); #plot our data
        plt.show() #show

        loss = np.sum((y_guess - y)**2) #this is the loss formula, the sum of the square difference between our y and the guess y line

        print(loss,'M = ',m_guess) #we print the guess for each and the loss as well.  By determining the lowest loss we determine the best possible value for M, however this method is extremely inefficient
```

```python
best_loss = 100000000000 #now we will make it slightly more efficient by only saving the best guess, not every single guess
best_m = 0 #our best slope will start as 0
best_b = 0 #our best y int will start as 0, these will be updated with each run if and only if the loss is lower than the previous best guess

for m_guess in np.linspace(-100,100,100): #now we will go with 100 possible slopes and 100 possible y ints, making a total of 10,000 guesses
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * x + b_guess #we will generate data with our guess, and see the difference between it and the real data

        loss = np.sum((y_guess - y)**2) #using the loss function we will update our best_loss, best_m, and best_b variables

        if loss < best_loss: #if the loss is lower than the previous best we update the data, otherwise we keep the previous data

            best_loss = loss #these only trigger if loss is lower than previous best
            best_m = m_guess
            best_b = b_guess
```

```python
y_guess = best_m * x + best_b #now we save our best y to a variable after acquiring our best b and best m with the best loss

plt.plot(x,y_guess,'r'); #plot our guess
plt.plot(x,y2,'g.'); #plot our data
plt.show() #show the plot

print(best_loss,best_m,best_b) #print our values
```

```python
x = D #x is our diameter
y = C #y is our circumference
```

```python
m_guess = 0 #now let's guess the values of m and b
b_guess = 0
```

```python
x #this is our x values
```

```python
m_guess * x #this is our m guess multiplied by x
```

```python
m_guess * x + b_guess #y=mx+b (linear function)
```

```python
y_guess = m_guess * x + b_guess #we want to find the line that best fits
```

```python
y_guess #by using our guess values for m and b we generate a line
```

```python
best_loss = 1000000000 #again we use the method of only updating the values in the system when the loss is lower than the previous best
best_m = 0 #these are our starting values for m and b, where we will save our best values when loss is lowest
best_b = 0

for m_guess in np.linspace(-100,100,100): #another 10000 guesses, 100 for each slope and y int
    for b_guess in np.linspace(-100,100,100):

        print("Now Testing: ", m_guess, b_guess) #this will print out the guesses one by one

        y_guess = m_guess * x + b_guess #this is our guess, which will not be saved after the function runs, thus saving computing time

        loss = np.sum((y_guess - y)**2) #we use this and compare against our known results to see how low the loss is

        if loss < best_loss: #if the loss is lower than the previous best we save the values of our guesses

            best_loss = loss
            best_m = m_guess
            best_b = b_guess
```

```python
y_guess = best_m * x + best_b #now we will find the line of best fit by using our best guess values

plt.plot(x,y_guess,'r'); #plot line of best fit
plt.plot(D,C,'g.'); #plot the actual data
plt.show() #show

print(best_loss,best_m,best_b) #show the values for best loss, best m and best b
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([19,65,66,80,97,113,175,270,310,350,350,650,680]) #with the redshift calculations we can find the distance of each galaxy and the speed at which it is moving away from us
Y = np.array([1150,3800,5000,4800,5400,6700,10300,15400,19500,23200,21600,39400,41000])
```

```python
plt.plot(X,Y,'g.') #plot the distance on the x axis and the velocity on the y
```

```python
m_guess = 0 #now we will guess the linear line of best fit
b_guess = 0
```

$y=mx+b$

```python
m_guess * X + b_guess #again we take our data and put our guess in to save it as y_guess
```

```python
y_guess = m_guess * X + b_guess
```

```python
y_guess - Y #we can see the difference between the guess and the real value
```

```python
np.sum((y_guess - Y)**2) #take the loss of these values summed and we get the loss of the whole function
```

```python
loss = np.sum((y_guess - Y)**2) #save it as a variable
```

```python
best_loss_so_far = 1000000000000 #best loss save variable
m_best = 0 #where we save the best guess
b_best = 0 #where we save the best y int
for m_guess in np.linspace(-100,100,100): #10,000 guesses to find the line of best fit
    for b_guess in np.linspace(-100,100,100):

        y_guess = m_guess * X + b_guess #find the y guess to compare it to the real y

        loss = np.sum((y_guess - Y)**2) #find the loss by comparing

        if loss < best_loss_so_far: #test loss vs previous best and if better save the m and b values
            m_best = m_guess
            b_best = b_guess
            best_loss_so_far = loss

print("Found best fit:") #print our best values and the loss
print(m_best,b_best,loss)
```

If the universe has been expanding since its beginning at a constant speed, the universe's age would simply be 1/Ho.
Find the inverse of your value of Ho.
Multiply the inverse by $3.09 \times 10^{19}$ km/Mpc to cancel the distance units.
Since you now have the age of the Universe in seconds, divide this number by the number of seconds in a year: $3.16 \times 10^{7}$ sec/yr.

$3.09 \times 10^{19}$ km/Mpc

$3.16 \times 10^{7}$ sec/year

```python
(1/H) * (3.09*10**19) #this changes the units to kilometers
```

```python
(1/H) * (3.09*10**19)  / (3.16*10**7) #this changes the units to seconds
```

```python
 (1/H) * (3.09*10**19)  / (3.16*10**7) / (1000000000) #this changes the units to per billion years
```

```python
import pandas as pd #import the pandas library
import numpy as np
import matplotlib.pyplot as plt
def get_database(url,sheet_name): #this function will get our data out of the sheet for use
    url_head = "https://docs.google.com/spreadsheets/d/"
    url_foot = "/gviz/tq?tqx=out:csv&sheet="
    url_body = url.split('/')[5]
    # sheet_name = "1"
    url_csv = url_head + url_body + url_foot + sheet_name
    df = pd.read_csv(url_csv, sep=',', skiprows=0)
    return df
```

```python
url = 'https://docs.google.com/spreadsheets/d/1B_HW2uVGQv9W7b6vVZDSkOPLtmfodHrX4G0qtQbIenk/edit?usp=sharing'
```
